```python
one=pd.DataFrame(t1)
two=pd.DataFrame(t2)
```

```python
pd.concat([one,two])
```

```python
# by defualt the axis = 0 
```

```python
pd.concat([one,two],axis=1)
```

```python
# to make the marge both dataframe in same columns , then needed to change the columns name in one of them
```

```python
two.columns=one.columns
```

```python
pd.concat([one,two],axis=0)
```

```python
# the table merge can be done by three main ways using how parameter
```

```python
reg = {'reg_id':[1,2,3,4],'name':['andrew','bob','charlie','david']}
log = {'log-id':[1,2,3,4],'name':['xavier','andrew','yolanda','bob']}
```

```python
reg=pd.DataFrame(reg)
log=pd.DataFrame(log)
```

```python
reg
```

```python
log
```

```python
# Inner merge
pd.merge(reg,log, how='inner',on='name')
```

```python
# Left or Righr merge
pd.merge(reg,log, how='left',on='name')
# left dataframe will be (reg) and all it contant will show up , while the right dataframe will not show all its content , only 
# the common with the left dataframe
```

```python
pd.merge(reg,log, how='right',on='name')
```

```python
# Outer merge
pd.merge(reg,log, how='outer',on='name')
```

```python
reg = reg.set_index('name')
```

```python
reg
```

```python
log
```

```python

pd.merge(reg,log, how='inner',left_index=True,right_on='name')
```

```python
reg = reg.reset_index()
```

```python
reg
```

```python
# in case the columns has different names 
reg.columns= ['person', 'reg-id']
```

```python
reg
```

```python
pd.merge(reg,log, left_on ='person', right_on='name', how='inner')
```

```python
# the column(person)& column(name) is the same so can drop any of them
```

```python
# in case that the columns has same names , then pandas will add suffix 
```

```python
reg.columns=['name','id']
log.columns=['id','name']
```

```python
reg
```

```python
log
```

```python
pd.merge(reg,log,how='inner',on='name')
```

```python
pd.merge(reg,log,how='inner',on='name',suffixes=('-reg','-log'))
```
